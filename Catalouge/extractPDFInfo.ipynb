```python
import pdfplumber
import os

def extract_pages_pdfplumber(pdf_path, output_folder="pdf_pages_plumber"):
    try:
        os.makedirs(output_folder, exist_ok=True)

        with pdfplumber.open(pdf_path) as pdf:
            total_pages = len(pdf.pages)
            print(f"Total pages: {total_pages}")

            for page_num, page in enumerate(pdf.pages, start=1):
                text = page.extract_text()

                pdf_name = os.path.splitext(os.path.basename(pdf_path))[0]
                filename = f"{pdf_name}_page_{page_num}.txt"
                filepath = os.path.join(output_folder, filename)

                if text and text.strip():
                    with open(filepath, 'w', encoding='utf-8') as f:
                        f.write(text.strip())
                    print(f"Saved page {page_num}")
                else:
                    print(f"Skipped empty page {page_num}")

        print("\nExtraction complete!")

    except Exception as e:
        print(f"Error: {e}")


# ðŸ”¥ YOUR PDF PATH
extract_pages_pdfplumber(r"D:\System_Data\Documents\GitHub\Dayita-Inventory\cat 1-100.pdf")
```

```python
import pytesseract
import os
from PIL import Image
import fitz
import re

# ===============================
# SETTINGS
# ===============================

PDF_PATH = r"D:\System_Data\Documents\GitHub\Dayita-Inventory\Catalouge\D&H Catalogue & Welding Handbook.pdf"
OUTPUT_FOLDER = "ai_catalogue_data"

pytesseract.pytesseract.tesseract_cmd = r"C:\Users\DNH\AppData\Local\Programs\Tesseract-OCR\tesseract.exe"

# ===============================
# STEP 1 â€” OCR EXTRACT
# ===============================

def extract_pdf_text():
    os.makedirs(OUTPUT_FOLDER, exist_ok=True)

    print("Opening catalogue...")
    doc = fitz.open(PDF_PATH)

    for page_num in range(len(doc)):
        print(f"OCR Reading page {page_num+1}...")

        page = doc.load_page(page_num)
        pix = page.get_pixmap(dpi=300)

        img_path = os.path.join(OUTPUT_FOLDER, f"page_{page_num+1}.png")
        pix.save(img_path)

        img = Image.open(img_path)
        text = pytesseract.image_to_string(img)

        txt_path = os.path.join(OUTPUT_FOLDER, f"page_{page_num+1}.txt")
        with open(txt_path, "w", encoding="utf-8") as f:
            f.write(text)

    print("OCR extraction done.")

# ===============================
# STEP 2 â€” CLEAN TEXT
# ===============================

def clean_text():
    print("Cleaning OCR text...")

    for file in os.listdir(OUTPUT_FOLDER):
        if file.endswith(".txt"):
            path = os.path.join(OUTPUT_FOLDER, file)

            with open(path, "r", encoding="utf-8") as f:
                text = f.read()

            # remove garbage
            text = re.sub(r'\n+', '\n', text)
            text = re.sub(r'[^\w\s\.\-\(\)%/]', '', text)

            # fix common OCR words
            text = text.replace("DjH", "D&H")
            text = text.replace("secheron", "Secheron")

            with open(path, "w", encoding="utf-8") as f:
                f.write(text)

    print("Cleaning completed.")

# ===============================
# STEP 3 â€” SEARCH ENGINE
# ===============================

def search_product(keyword):
    print(f"\nSearching for: {keyword}\n")

    results = []

    for file in os.listdir(OUTPUT_FOLDER):
        if file.endswith(".txt"):
            path = os.path.join(OUTPUT_FOLDER, file)

            with open(path, "r", encoding="utf-8") as f:
                text = f.read()

            if keyword.lower() in text.lower():
                results.append((file, text[:1000]))

    if results:
        for r in results:
            print("Found in:", r[0])
            print(r[1])
            print("\n-------------------\n")
    else:
        print("No product found.")

# ===============================
# MAIN CONTROL
# ===============================

if __name__ == "__main__":

    print("""
    1 - Extract Catalogue OCR
    2 - Clean OCR Text
    3 - Search Product
    """)

    choice = input("Enter option: ")

    if choice == "1":
        extract_pdf_text()

    elif choice == "2":
        clean_text()

    elif choice == "3":
        keyword = input("Enter product name: ")
        search_product(keyword)
```

```python
#converting into json

import os
import json

OCR_FOLDER = r"D:\System_Data\Documents\GitHub\Dayita-Inventory\Catalouge\catalogue_data"
OUTPUT_JSON = r"D:\System_Data\Documents\GitHub\Dayita-Inventory\products_ai.json"

products = []

for file in os.listdir(OCR_FOLDER):
    if file.endswith(".txt"):
        path = os.path.join(OCR_FOLDER, file)

        with open(path, "r", encoding="utf-8") as f:
            text = f.read()

        # detect product name (simple logic)
        lines = text.split("\n")
        name = ""
        for l in lines[:10]:
            if len(l.strip()) > 4 and l.isupper():
                name = l.strip()
                break

        products.append({
            "product_name": name,
            "page": file,
            "content": text
        })

with open(OUTPUT_JSON, "w", encoding="utf-8") as f:
    json.dump(products, f, indent=2)

print("AI JSON DATABASE CREATED")
```
